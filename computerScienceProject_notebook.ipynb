Raccolgo in un unico dataframe tutte le università più recenti con tutti i ranking

```python
tot_most_year = completeMerge[['institution', 'year_cwur','world_rank_cwur', 'year_times','world_rank_times', 'year', 'world_rank_shanghai']].copy()

#uso il datetime per far riconoscere le colonne seguenti come anni
tot_most_year['year_times'] = pd.to_datetime(tot_most_year['year_times'], format = '%Y')
tot_most_year['year'] = pd.to_datetime(tot_most_year['year'], format = '%Y')
tot_most_year['year_cwur'] = pd.to_datetime(tot_most_year['year_cwur'], format = '%Y')

#splitto il ranking dividenndo in un rank minimo e uno massimo
datiSplit = tot_most_year['world_rank_times'].str.split('-', expand=True)
tot_most_year[['rankMinT', 'rankMaxT']] = datiSplit

datiSplit2 = tot_most_year['world_rank_shanghai'].str.split('-', expand=True)
tot_most_year[['rankMinS', 'rankMaxS']] = datiSplit2

#trasformo in float i valori dei ranking
tot_most_year['rankMinT'] = tot_most_year['rankMinT'].astype(float).copy()
tot_most_year['rankMaxT'] = tot_most_year['rankMaxT'].astype(float).copy()
tot_most_year['rankMinS'] = tot_most_year['rankMinS'].astype(float).copy()
tot_most_year['rankMaxS'] = tot_most_year['rankMaxS'].astype(float).copy()


tot_most_year
```

```python
# inserisco il valore max dei ranking di quella riga in max_value, faccio la stessa cosa per il minimo in min_value
tot_most_year['max_value'] = tot_most_year.max(axis=1)
tot_most_year
tot_most_year['min_value'] = tot_most_year[['world_rank_cwur','rankMinT', 'rankMinS']].min(axis=1)
tot_most_year
```

Uso una funzione "comparaDifference" che controlla per prima cosa se il valore in max_value e in min_value è uguale, in caso affermativo significa che si ha un solo valore tra tutti i ranking a disposizione, quindi la funzione restituisce tale valore.
Altrimenti la funzione calcola tutte le combinazioni di differenze possibili ed inserisce dentro la colonna "MaxDifference" il valore più grande della differenza calcolata.

```python
def comparaDifferenze(riga):
    
    if (riga['max_value'] == riga['min_value']):
        return riga['max_value'] 
    
    a = riga['world_rank_cwur'] - riga['rankMinT'] 
    b = riga['world_rank_cwur'] - riga['rankMaxT']
    c = riga['world_rank_cwur'] - riga['rankMinS']  
    d = riga['world_rank_cwur'] - riga['rankMaxS']
    
    e = riga['rankMinT'] - riga['world_rank_cwur'] 
    f = riga['rankMinT'] - riga['rankMinS']
    g = riga['rankMinT'] - riga['rankMaxS']  

    h = riga['rankMaxT'] - riga['world_rank_cwur'] 
    i = riga['rankMaxT'] - riga['rankMinS']
    j = riga['rankMaxT'] - riga['rankMaxS']
    
    k = riga['rankMinS'] - riga['world_rank_cwur'] 
    l = riga['rankMinS'] - riga['rankMinT']
    m = riga['rankMinS'] - riga['rankMaxT']
    
    n = riga['rankMaxS'] - riga['world_rank_cwur'] 
    o = riga['rankMaxS'] - riga['rankMinT']
    p = riga['rankMaxS'] - riga['rankMaxT']
    
    MaxDif = max(a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p)
    return MaxDif
```

```python
tot_most_year['maxDifference'] = tot_most_year.apply(comparaDifferenze, axis = 1)
```

## 10. For each university, find the number of rankings in which they appear (it suffices to appear in one year for each ranking).

```python
#conto le volte che ogni università è dentro un ranking
timesRank = times.groupby('university_name')['world_rank'].count().reset_index()
cwurRank = cwurData.groupby('institution')['world_rank'].count().reset_index()
shanghaiRank = shanghaiData.groupby('university_name')['world_rank'].count().reset_index()
#rinomino la colonna di cwurRank
cwurRank.rename(columns={'institution': 'university_name'}, inplace=True)
```

```python
#unisco insieme le tabelle
merge1 = pd.merge(cwurRank, shanghaiRank, left_on='university_name', right_on='university_name', how='left', suffixes=('_cwur','_shanghai'))
totalMergeRank = pd.merge(merge1, timesRank, left_on='university_name', right_on='university_name', how='left', suffixes=('_shanghai', '_times'))

#rinomino le colonne 
totalMergeRank.rename(columns={'world_rank_cwur': 'appearances_cwur'}, inplace=True)
totalMergeRank.rename(columns={'world_rank': 'appearances_times'}, inplace=True)
totalMergeRank.rename(columns={'world_rank_shanghai': 'appearances_shanghai'}, inplace=True)

totalMergeRank
```

```python
# creo un dizionario con i nomi delle università ed un contatore per ognuno di essi uguale a zero

numero_apparizioni = {}
for university in totalMergeRank['university_name']:
    numero_apparizioni[university] = 0

# funzione che aumenta il contatore per una università di uno ogni per ogni ranking in cui appare

def number_rankings(riga):
    if riga['appearances_cwur'] >= 1:
        numero_apparizioni[riga['university_name']] +=1
    if riga['appearances_shanghai'] >= 1:
        numero_apparizioni[riga['university_name']] +=1
    if riga['appearances_times'] >= 1:
        numero_apparizioni[riga['university_name']] +=1
    return(numero_apparizioni)
```

```python
totalMergeRank.apply(number_rankings, axis=1)
numero_apparizioni
```
